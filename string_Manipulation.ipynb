```python
'''WAP to check whether the entered password is valid or not.
Length of password >= 8
Atleast one special symbol
Atleast one upper case character
Atleast one digit'''

import string
x = string.punctuation
 
a = input()
b = 0
c= 0
d = 0
e =0
f =0

for i in a:
  if i.isupper() == True:
    b =1
  if i.islower() == True:
    c =1
  if i.isdigit() == True:
    d =1
  if i.isnumeric == True:
    e =1
  if i in x :
    f = 1

if b == 1 and c == 1 and d == 1 and e == 1 and f == 1:
  print('its a valid password')

else :
  print('its not a valid password')
```
