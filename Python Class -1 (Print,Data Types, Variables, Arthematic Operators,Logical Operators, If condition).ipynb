### 1. Simple **if**

keyword - **if**

Block of code -> indented

```
Syntax :-   if (condition):
                Statement 1 
                Statement 2
                    .
                    .
                Statement n
            Statement x
```



### 2. **if** - **else**

keywords - **if** & **else**

Two Blocks of code -> indented

**if** block = True block

**else** block = False block

```
Syntax :-   if (condition):
                Statement 1 
                Statement 2
                    .
                    .
                Statement n
            else:
                Statement 1 
                Statement 2
                    .
                    .
                Statement n
            Statement x
```


### 3. Nested **if** - **else**

keywords - **if** & **else**

Multiple Blocks of code -> indented

**if** block = True block

**else** block = False block

Used when **multiple dependent conditions** are present

```
Syntax :-   if (condition 1):
                # True Block
                if (condition 2):
                   # Inner True block
                   Statement 1 
                   Statement 2
                else:
                   # Inner False block
                   Statement 1 
                   Statement 2
            else:
                # False block
                if (condition 3):
                   # Inner True block
                   Statement 1 
                   Statement 2
                else:
                   # Inner False block
                   Statement 1 
                   Statement 2
            Statement x
```


### 4. **if** - **elif** - **else**

keywords - **if** , **elif**, **else**

Multiple Blocks of code -> indented

**if** block = True block

**elif** block = True block

**else** block = False block

Used when **multiple independent conditions** are present 

```
Syntax :-   if (condition 1):
                Statement 1 
                Statement 2
            elif (condition 2):
                Statement 1 
                Statement 2
                    .
                    .
            else:
                Statement 1
                Statement 2
            Statement x
```
